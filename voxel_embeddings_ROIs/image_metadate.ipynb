Testing saving images with metadata

```python
import os 
from PIL import Image, PngImagePlugin
from IPython.display import display



path = '/home/matanyaw/DIP_decoder/matanya_results/results_25_07_21/run_test_metadata_long2/img_21/roi_FBA-1/img_21_stroke_Top-K_ROI_FBA-1_start_fMRI.png'
img = Image.open(path)
display(img)
for key in img.info.keys():
    print(f"{key}: {img.info[key]}")


```

```python
pnginfo = PngImagePlugin.PngInfo()
for k, v in img.info.items():
    pnginfo.add_text(k, str(v))
pnginfo.add_text('title', 'Top-K start fMRI')

# for k, v in pnginfo.text.items():
#     print(f"{k}: {v}")
```

```python
img.save(path, pnginfo=pnginfo)
```
